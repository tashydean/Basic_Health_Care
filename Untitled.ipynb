```python
import os
import pandas as pd
import requests

# ==========================================
# ÏÑ§Ï†ï: Î°úÏª¨ÏóêÏÑú ÏûëÏóÖÌï† ÎïåÎäî True, ÍπÉÌóàÎ∏åÏóêÏÑú Î∂àÎü¨Ïò¨ ÎïåÎäî False
# ==========================================
LOCAL_MODE = True  
DATA_DIR = "./Data_Files"
os.makedirs(DATA_DIR, exist_ok=True)

URLS = {
    "AdmissionsCorePopulatedTable": "https://raw.githubusercontent.com/tashydean/Basic_Health_Care/refs/heads/master/Data_Files/AdmissionsCorePopulatedTable.txt",
    "AdmissionsDiagnosesCorePopulatedTable": "https://raw.githubusercontent.com/tashydean/Basic_Health_Care/refs/heads/master/Data_Files/AdmissionsDiagnosesCorePopulatedTable.txt",
    "LabsCorePopulatedTable": "https://raw.githubusercontent.com/tashydean/Basic_Health_Care/refs/heads/master/Data_Files/LabsCorePopulatedTable.txt",
    "PatientCorePopulatedTable": "https://raw.githubusercontent.com/tashydean/Basic_Health_Care/refs/heads/master/Data_Files/PatientCorePopulatedTable.txt"
}

# ---------------------------
# Ïú†Ìã∏Î¶¨Ìã∞ Ìï®ÏàòÎì§
# ---------------------------
def download_to_path(url, dest_path):
    """URLÏóêÏÑú ÌååÏùºÏùÑ Îã§Ïö¥Î°úÎìú."""
    print(f"Downloading from GitHub -> {dest_path}...")
    r = requests.get(url)
    r.raise_for_status()
    with open(dest_path, 'wb') as f:
        f.write(r.content)

def read_table_with_schema(path, schema):
    """ÏßÄÏ†ïÌïú Ïä§ÌÇ§ÎßàÎ°ú ÌååÏùº ÏùΩÍ∏∞."""
    df = pd.read_csv(path, sep='\t', skiprows=1, header=None, names=schema['names'], dtype=str)
    if schema.get('coerce_numeric'):
        for c in schema['coerce_numeric']:
            df[c] = pd.to_numeric(df[c], errors='coerce')
    return df

def load_all_data(mode_local=True):
    """Î°úÏª¨ ÌòπÏùÄ Ïò®ÎùºÏù∏ÏóêÏÑú Îç∞Ïù¥ÌÑ∞Î•º Î°úÎìúÌïòÍ≥† Ï†ÑÏó≠ Î≥ÄÏàò ÏÉùÏÑ±."""
    data_frames = {}
    schemas = {
        "AdmissionsCorePopulatedTable": {"names": ["PatientID","AdmissionID","AdmissionStartDate","AdmissionEndDate"], "coerce_numeric": []},
        "AdmissionsDiagnosesCorePopulatedTable": {"names": ["PatientID","AdmissionID","PrimaryDiagnosisCode","PrimaryDiagnosisDescription"], "coerce_numeric": []},
        "LabsCorePopulatedTable": {"names": ["PatientID","AdmissionID","LabName","LabValue","LabUnits","LabDateTime"], "coerce_numeric": ["LabValue"]},
        "PatientCorePopulatedTable": {"names": ["PatientID","PatientGender","PatientDateOfBirth","PatientRace","PatientMaritalStatus","PatientLanguage","PatientPopulationPercentageBelowPoverty"], "coerce_numeric": ["PatientPopulationPercentageBelowPoverty"]}
    }
    
    mapping = {
        "AdmissionsCorePopulatedTable": "df_ACPT",
        "AdmissionsDiagnosesCorePopulatedTable": "df_ADCPT",
        "LabsCorePopulatedTable": "df_LCPT",
        "PatientCorePopulatedTable": "df_PCPT"
    }

    for key, url in URLS.items():
        local_path = os.path.join(DATA_DIR, f"{key}.txt")
        
        if mode_local:
            # Î°úÏª¨ Î™®Îìú: ÌååÏùºÏù¥ ÏûàÎäîÏßÄ ÌôïÏù∏Îßå Ìï®
            if not os.path.exists(local_path):
                print(f"‚ö†Ô∏è Î°úÏª¨ ÌååÏùºÏù¥ ÏóÜÏäµÎãàÎã§. Îã§Ïö¥Î°úÎìúÎ•º ÏãúÎèÑÌï©ÎãàÎã§: {key}")
                download_to_path(url, local_path)
        else:
            # Ïò®ÎùºÏù∏ Î™®Îìú: Ìï≠ÏÉÅ Îã§Ïö¥Î°úÎìú (ÌòπÏùÄ ÎçÆÏñ¥Ïì∞Í∏∞)
            download_to_path(url, local_path)
            
        df = read_table_with_schema(local_path, schemas[key])
        data_frames[key] = df
        globals()[mapping[key]] = df  # Ï†ÑÏó≠ Î≥ÄÏàò ÏÉùÏÑ±
        
    print("‚úÖ Î™®Îì† Îç∞Ïù¥ÌÑ∞ Î°úÎìú ÏôÑÎ£å!")
    return data_frames

# Îç∞Ïù¥ÌÑ∞ Î°úÎìú Ïã§Ìñâ
data_frames = load_all_data(mode_local=LOCAL_MODE)
```

```python

# ---------------------------
# Í≥µÌÜµ Ï†ÑÏ≤òÎ¶¨ (ÎÇ†Ïßú, Ïπ¥ÌÖåÍ≥†Î¶¨, Ïà´Ïûê Îì±)
# ---------------------------
def common_preprocess(dfs):
    """ÎÇ†ÏßúÌòï Î≥ÄÌôò, Î≤îÏ£ºÌòï Î≥ÄÌôò, Ïà´Ïûê Î≥ÄÌôòÏùÑ ÏùºÍ¥Ñ Ï≤òÎ¶¨."""
    date_cols = ['LabDateTime','PatientDateOfBirth','AdmissionStartDate','AdmissionEndDate']
    for key, df in dfs.items():
        for col in date_cols:
            if col in df.columns:
                df[col] = pd.to_datetime(df[col], errors='coerce')
    # AdmissionID Î≤îÏ£ºÌòï(ordered)
    for key, df in dfs.items():
        if 'AdmissionID' in df.columns:
            df['AdmissionID'] = df['AdmissionID'].astype('category').cat.as_ordered()
    # Ïù∏Íµ¨ÌÜµÍ≥Ñ Î≤îÏ£ºÌòï
    cat_cols = ['PatientGender','PatientRace','PatientMaritalStatus','PatientLanguage','PrimaryDiagnosisCode','LabName']
    for key, df in dfs.items():
        for col in cat_cols:
            if col in df.columns:
                df[col] = df[col].astype('category')
    # ÎπàÍ≥§Ïú® Ïà´Ïûê Î≥ÄÌôò
    for key, df in dfs.items():
        if 'PatientPopulationPercentageBelowPoverty' in df.columns:
            df['PatientPopulationPercentageBelowPoverty'] = pd.to_numeric(df['PatientPopulationPercentageBelowPoverty'], errors='coerce')

common_preprocess(data_frames)

# Í∞Å DataFrameÏùÑ Ï†ÑÏó≠ Î≥ÄÏàòÎ°ú Îã§Ïãú Î∞îÏù∏Îî© (Ìé∏Ïùò)
df_PCPT = globals().get('df_PCPT')
df_ACPT = globals().get('df_ACPT')
df_ADCPT = globals().get('df_ADCPT')
df_LCPT = globals().get('df_LCPT')

# ---------------------------
# ÏùºÎ∞òÏ†ÅÏù∏ EDA Ïú†Ìã∏ (ÏõêÎ≥∏ ÏΩîÎìúÏùò Ï∂úÎ†• Ìè¨Îß∑ÏùÑ Ïú†ÏßÄ)
# ---------------------------
def general_overview(df, name):
    print(f"\n------------------------ {name} Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ Í∏∞Ï¥à Î∂ÑÏÑù ÏãúÏûë ------------------------")
    # 1. Îç∞Ïù¥ÌÑ∞ ÌÉÄÏûÖ
    print(df.dtypes)
    # 2. Í≤∞Ï∏°Ïπò
    print("Í≤∞Ï∏°Ïπò Í∞úÏàò : ")
    print(df.isnull().sum())
    # 3. head, describe, Ï§ëÎ≥µ Îì±
    print(df.head())
    print(df.describe(include='all', datetime_is_numeric=True))
    print('Ï§ëÎ≥µÎêú Ìñâ Ïàò:', df.duplicated().sum())
    print('Ï§ëÎ≥µ Ìñâ ÏòàÏãú : ')
    if df.duplicated().sum() > 0:
        print(df[df.duplicated()].head())
    else:
        print("Ï§ëÎ≥µ ÏóÜÏùå")

# ÏõêÎ≥∏ÏóêÏÑú ÌñàÎçò Í∏∞Î≥∏ EDA (Î™®Îì† dfÏóê ÎåÄÌï¥)
for key, df in data_frames.items():
    general_overview(df, key)
```

```python
# =============================================================================
# ÏùòÎ£å AI Ìè¨Ìä∏Ìè¥Î¶¨Ïò§ - Ïã§Ìå®Î•º ÏûêÏÇ∞ÏúºÎ°ú ÎßåÎìúÎäî Î∂ÑÏÑù
# "ÏûëÏùÄ ÏÉòÌîåÏóêÏÑú ÏùòÎØ∏ÏûàÎäî Ïù∏ÏÇ¨Ïù¥Ìä∏ Ï∞æÍ∏∞"
# =============================================================================

import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score
from scipy import stats


# =============================================================================
# PART 0: ÏõêÎ≥∏ ÏΩîÎìúÏùò ÌïµÏã¨ Ìï®Ïàò Ïû¨Íµ¨ÌòÑ (ÏùòÎ£åÏ†Å ÏùòÎØ∏ Î≥¥Ï°¥)
# =============================================================================

def detect_abnormal_patients_v2(master, stat_estimate=1.5):
    """
    ÏõêÎ≥∏ ÏΩîÎìúÏùò abnormal_behavior_detection Í∞úÏÑ† Î≤ÑÏ†Ñ
    - Î™©Ï†Å: Ïù¥ÏÉÅ ÏàòÏπòÎ•º Í∞ÄÏßÑ "ÌôòÏûê"Î•º Ï∞æÎäîÎã§ (Í≤ÄÏÇ¨ Îã®ÏúÑÍ∞Ä ÏïÑÎãò)
    - ÏùòÎ£åÏ†Å ÏùòÎØ∏: Ï†ïÏÉÅ Î≤îÏúÑ Î≤óÏñ¥ÎÇú ÌôòÏûê = Ï§ëÏ¶ùÎèÑ ÎÜíÏùå
    """
    print("="*70)
    print("üî¨ Ïù¥ÏÉÅ ÏàòÏπò ÌôòÏûê ÌÉêÏßÄ (ÏõêÎ≥∏ Î°úÏßÅ Í∏∞Î∞ò)")
    print("="*70)
    
    # 1. Í≤ÄÏÇ¨Î≥Ñ Ï†ïÏÉÅ Î≤îÏúÑ Í≥ÑÏÇ∞ (ÌèâÍ∑† ¬± 1.5*ÌëúÏ§ÄÌé∏Ï∞®)
    lab_stats = master.groupby('LabName')['LabValue'].agg(['mean', 'std']).reset_index()
    lab_stats['upper'] = lab_stats['mean'] + stat_estimate * lab_stats['std']
    lab_stats['lower'] = lab_stats['mean'] - stat_estimate * lab_stats['std']
    
    # 2. Í∞Å Í≤ÄÏÇ¨ Í≤∞Í≥ºÍ∞Ä Ï†ïÏÉÅ Î≤îÏúÑÎ•º Î≤óÏñ¥ÎÇ¨ÎäîÏßÄ ÌåêÎã®
    master_check = master.merge(lab_stats[['LabName', 'upper', 'lower']], on='LabName', how='left')
    master_check['IsAbnormal'] = (
        (master_check['LabValue'] > master_check['upper']) | 
        (master_check['LabValue'] < master_check['lower'])
    ).astype(int)
    
    # 3. ÌôòÏûêÎ≥Ñ Ïù¥ÏÉÅ ÏàòÏπò Í∞úÏàò ÏßëÍ≥Ñ
    patient_abnormal_count = master_check.groupby('PatientID').agg({
        'IsAbnormal': 'sum',
        'LabValue': 'count'
    }).reset_index()
    patient_abnormal_count.columns = ['PatientID', 'AbnormalCount', 'TotalTests']
    patient_abnormal_count['AbnormalRatio'] = (
        patient_abnormal_count['AbnormalCount'] / patient_abnormal_count['TotalTests']
    )
    
    # 4. Í≥†ÏúÑÌóò ÌôòÏûê Ï†ïÏùò: Ïù¥ÏÉÅ ÏàòÏπò ÎπÑÏú® > 50%
    high_risk_patients = patient_abnormal_count[patient_abnormal_count['AbnormalRatio'] > 0.5]
    
    print(f"‚úÖ Ï†ÑÏ≤¥ ÌôòÏûê: {len(patient_abnormal_count)}Î™Ö")
    print(f"üö® Í≥†ÏúÑÌóò ÌôòÏûê: {len(high_risk_patients)}Î™Ö ({len(high_risk_patients)/len(patient_abnormal_count)*100:.1f}%)")
    print(f"ÌèâÍ∑† Ïù¥ÏÉÅ ÏàòÏπò ÎπÑÏú®: {patient_abnormal_count['AbnormalRatio'].mean()*100:.1f}%")
    
    # ÏãúÍ∞ÅÌôî
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    
    axes[0].hist(patient_abnormal_count['AbnormalRatio'], bins=20, edgecolor='black', alpha=0.7)
    axes[0].axvline(x=0.5, color='red', linestyle='--', label='Í≥†ÏúÑÌóò Í∏∞Ï§Ä (50%)')
    axes[0].set_title('ÌôòÏûêÎ≥Ñ Ïù¥ÏÉÅ ÏàòÏπò ÎπÑÏú® Î∂ÑÌè¨')
    axes[0].set_xlabel('Ïù¥ÏÉÅ ÏàòÏπò ÎπÑÏú®')
    axes[0].set_ylabel('ÌôòÏûê Ïàò')
    axes[0].legend()
    
    axes[1].scatter(patient_abnormal_count['TotalTests'], 
                   patient_abnormal_count['AbnormalRatio'], 
                   alpha=0.6, s=50)
    axes[1].set_title('Í≤ÄÏÇ¨ Ïàò vs Ïù¥ÏÉÅ ÏàòÏπò ÎπÑÏú®')
    axes[1].set_xlabel('Ï¥ù Í≤ÄÏÇ¨ Ïàò')
    axes[1].set_ylabel('Ïù¥ÏÉÅ ÏàòÏπò ÎπÑÏú®')
    axes[1].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    return patient_abnormal_count, high_risk_patients


# ÏõêÎ≥∏ Master Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ± (ÏõêÎ≥∏ ÏΩîÎìú Î∞©Ïãù)
def create_original_master(admissions, diagnoses, labs, patients):
    """ÏõêÎ≥∏ ÏΩîÎìúÏùò create_master_dataset Í∑∏ÎåÄÎ°ú Ïû¨ÌòÑ"""
    patient_admission = pd.merge(patients, admissions, on='PatientID', how='left')
    patient_admission_diag = pd.merge(patient_admission, diagnoses, on=['PatientID', 'AdmissionID'], how='left')
    master = pd.merge(labs, patient_admission_diag, on=['PatientID', 'AdmissionID'], how='left')
    
    # ÎÇ†Ïßú Î≥ÄÌôò
    for col in ['LabDateTime', 'PatientDateOfBirth', 'AdmissionStartDate', 'AdmissionEndDate']:
        if col in master.columns:
            master[col] = pd.to_datetime(master[col], errors='coerce')
    
    return master


master_original = create_original_master(df_ACPT, df_ADCPT, df_LCPT, df_PCPT)
patient_abnormal, high_risk_list = detect_abnormal_patients_v2(master_original)


# =============================================================================
# PART 1: ÏßÑÎã®Î™Ö Î∂ÑÎ•ò Í∞úÏÑ† - ICD Í∏∞Î∞ò Body System
# =============================================================================

def classify_by_body_system_improved(desc):
    """ÏùòÌïôÏ†ÅÏúºÎ°ú ÏùòÎØ∏ÏûàÎäî Î∂ÑÎ•ò (ICD-10 Chapter Í∏∞Ï§Ä)"""
    desc = str(desc).lower()
    
    # ÌòàÏï°/Ï°∞Ìòà (D50-D89)
    if any(x in desc for x in ['blood', 'anemia', 'leuk', 'lymph', 'coagulation', 'hemoglobin']):
        return 'Hematologic'
    
    # ÏàúÌôòÍ∏∞ (I00-I99)
    elif any(x in desc for x in ['heart', 'cardiac', 'vascular', 'hypertens', 'coronary', 'arrhythmia']):
        return 'Circulatory'
    
    # Ìò∏Ìù°Í∏∞ (J00-J99)
    elif any(x in desc for x in ['lung', 'respiratory', 'pneumonia', 'asthma', 'copd', 'bronch']):
        return 'Respiratory'
    
    # ÎÇ¥Î∂ÑÎπÑ/ÎåÄÏÇ¨ (E00-E90)
    elif any(x in desc for x in ['diabet', 'thyroid', 'metabolic', 'glucose', 'obesity', 'endocrine']):
        return 'Endocrine'
    
    # Ïã†Ïû•/ÎπÑÎá®Í∏∞ (N00-N99)
    elif any(x in desc for x in ['renal', 'kidney', 'urinary', 'bladder', 'nephro']):
        return 'Genitourinary'
    
    # Í∞êÏóº (A00-B99)
    elif any(x in desc for x in ['infection', 'sepsis', 'bacteremia', 'viral', 'fever']):
        return 'Infectious'
    
    # Ïã†ÏÉùÎ¨º (C00-D48)
    elif any(x in desc for x in ['cancer', 'carcinoma', 'tumor', 'neoplasm', 'malignancy', 'lymphoma']):
        return 'Neoplasm'
    
    # ÏÜåÌôîÍ∏∞ (K00-K93)
    elif any(x in desc for x in ['gastro', 'liver', 'intestin', 'colon', 'digest', 'hepat']):
        return 'Digestive'
    
    else:
        return 'Other'


# Í∏∞Ï°¥ df_finalÏóê Ï†ÅÏö©
df_final['BodySystem'] = df_final['PrimaryDiagnosisDescription'].apply(classify_by_body_system_improved)

print("\n" + "="*70)
print("üè• Í∞úÏÑ†Îêú ÏßÑÎã® Î∂ÑÎ•ò (Body System)")
print("="*70)
print(df_final['BodySystem'].value_counts())

# ÏãúÍ∞ÅÌôî
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

df_final['BodySystem'].value_counts().plot(kind='bar', ax=axes[0], color='steelblue')
axes[0].set_title('ÏßÑÎã® Î∂ÑÌè¨ (Body System Í∏∞Ï§Ä)')
axes[0].set_xlabel('Body System')
axes[0].set_ylabel('ÌôòÏûê Ïàò')
axes[0].tick_params(axis='x', rotation=45)

# Í∏∞Ï°¥ Î∂ÑÎ•òÏôÄ ÎπÑÍµê
comparison = pd.DataFrame({
    'Í∏∞Ï°¥Î∂ÑÎ•ò': df_final['TargetCategory'].value_counts(),
    'Í∞úÏÑ†Î∂ÑÎ•ò': df_final['BodySystem'].value_counts()
}).fillna(0)
comparison.plot(kind='bar', ax=axes[1])
axes[1].set_title('Î∂ÑÎ•ò Î∞©Ïãù ÎπÑÍµê')
axes[1].set_ylabel('ÌôòÏûê Ïàò')
axes[1].tick_params(axis='x', rotation=45)
axes[1].legend(title='Î∂ÑÎ•ò Î∞©Ïãù')

plt.tight_layout()
plt.show()

print(f"\nüìä Î∂ÑÎ•ò Í∞úÏÑ† Ìö®Í≥º:")
print(f"- Í∏∞Ï°¥ 'Others' ÎπÑÏú®: {(df_final['TargetCategory']=='Others').sum()/len(df_final)*100:.1f}%")
print(f"- Í∞úÏÑ† 'Other' ÎπÑÏú®: {(df_final['BodySystem']=='Other').sum()/len(df_final)*100:.1f}%")


# =============================================================================
# PART 2: Íµ∞ÏßëÌôî Ïû¨Ìï¥ÏÑù - "Ìå®ÌÑ¥ Î∂ÄÏû¨"Î•º Ïù∏ÏÇ¨Ïù¥Ìä∏Î°ú
# =============================================================================

def meaningful_clustering_analysis(df_final):
    """
    Íµ∞ÏßëÌôî Ïã§Ìå®Î•º Ïù∏ÏÇ¨Ïù¥Ìä∏Î°ú Ï†ÑÌôò
    - Î™©Ï†Å: ÌôòÏûê Í∞Ñ ÏÉùÎ¶¨ÌïôÏ†Å Ìå®ÌÑ¥Ïù¥ ÏóÜÏùåÏùÑ ÏûÖÏ¶ù
    """
    print("\n" + "="*70)
    print("üî¨ Íµ∞ÏßëÌôî Î∂ÑÏÑù: Ìå®ÌÑ¥ ÌÉêÏßÄ Ïã§Ìå®Ïùò ÏùòÎØ∏")
    print("="*70)
    
    # Feature Ï§ÄÎπÑ
    lab_value_cols = [c for c in df_final.columns if c.startswith('LabValue_')]
    missing_ratio = df_final[lab_value_cols].isnull().mean()
    valid_cols = missing_ratio[missing_ratio < 0.5].index.tolist()
    
    print(f"ÏÇ¨Ïö© Í∞ÄÎä•Ìïú Í≤ÄÏÇ¨ Ìï≠Î™©: {len(valid_cols)}Í∞ú")
    
    X = df_final[valid_cols].copy()
    imputer = SimpleImputer(strategy='median')
    X_imputed = imputer.fit_transform(X)
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X_imputed)
    
    # Íµ∞ÏßëÌôî (K=3ÏúºÎ°ú Îã®ÏàúÌôî)
    kmeans = KMeans(n_clusters=3, random_state=42, n_init=30)
    df_final['Cluster'] = kmeans.fit_predict(X_scaled)
    sil_score = silhouette_score(X_scaled, df_final['Cluster'])
    
    print(f"Silhouette Score: {sil_score:.3f} (0.2 Ïù¥Ìïò = ÏïΩÌïú Ìå®ÌÑ¥)")
    print(f"\nÍµ∞ÏßëÎ≥Ñ ÌôòÏûê Ïàò:")
    print(df_final['Cluster'].value_counts())
    
    # Íµ∞Ïßë vs ÏßÑÎã® Í¥ÄÍ≥Ñ
    ct = pd.crosstab(df_final['Cluster'], df_final['BodySystem'], normalize='index')
    
    print(f"\nÍµ∞ÏßëÎ≥Ñ ÏßÑÎã® Î∂ÑÌè¨ (ÎπÑÏú®):")
    print(ct.round(2))
    
    # ÏãúÍ∞ÅÌôî
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    
    # 1. Íµ∞ÏßëÎ≥Ñ ÏßÑÎã® Î∂ÑÌè¨
    ct.plot(kind='bar', stacked=True, ax=axes[0])
    axes[0].set_title('Íµ∞ÏßëÎ≥Ñ ÏßÑÎã® Î∂ÑÌè¨ (Ìå®ÌÑ¥ Î∂ÄÏû¨ ÌôïÏù∏)')
    axes[0].set_xlabel('Cluster')
    axes[0].set_ylabel('ÎπÑÏú®')
    axes[0].legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)
    
    # 2. Íµ∞ÏßëÎ≥Ñ ÎÇòÏù¥ Î∂ÑÌè¨
    df_final.boxplot(column='AgeAtAdmission', by='Cluster', ax=axes[1])
    axes[1].set_title('Íµ∞ÏßëÎ≥Ñ ÎÇòÏù¥ Î∂ÑÌè¨')
    axes[1].set_xlabel('Cluster')
    plt.sca(axes[1])
    plt.xticks([1,2,3], ['Cluster 0', 'Cluster 1', 'Cluster 2'])
    
    # 3. Íµ∞ÏßëÎ≥Ñ ÏûÖÏõê Í∏∞Í∞Ñ
    if 'LengthOfStay_Days' in df_final.columns:
        df_final.boxplot(column='LengthOfStay_Days', by='Cluster', ax=axes[2])
        axes[2].set_title('Íµ∞ÏßëÎ≥Ñ ÏûÖÏõê Í∏∞Í∞Ñ')
        axes[2].set_xlabel('Cluster')
        plt.sca(axes[2])
        plt.xticks([1,2,3], ['Cluster 0', 'Cluster 1', 'Cluster 2'])
    
    plt.tight_layout()
    plt.show()
    
    # Ïù∏ÏÇ¨Ïù¥Ìä∏ ÎèÑÏ∂ú
    print("\n" + "="*70)
    print("üí° Íµ∞ÏßëÌôî Î∂ÑÏÑù Ïù∏ÏÇ¨Ïù¥Ìä∏")
    print("="*70)
    print(f"""
‚úÖ Ï£ºÏöî Î∞úÍ≤¨:
1. ÎÇÆÏùÄ Silhouette Score ({sil_score:.3f}) ‚Üí ÎöúÎ†∑Ìïú ÌôòÏûê Í∑∏Î£π ÏóÜÏùå
2. Íµ∞ÏßëÎ≥Ñ ÏßÑÎã® Î∂ÑÌè¨Í∞Ä Í±∞Ïùò Í∑†Îì± ‚Üí Í≤ÄÏÇ¨ ÏàòÏπòÏôÄ ÏßàÎ≥ë Í∞Ñ ÏßÅÏ†ëÏ†Å Ïó∞Í¥ÄÏÑ± ÏïΩÌï®
3. ÎÇòÏù¥/ÏûÖÏõê Í∏∞Í∞ÑÎèÑ Íµ∞Ïßë Í∞Ñ Ï∞®Ïù¥ ÎØ∏ÎØ∏

üìä ÏûÑÏÉÅÏ†Å Ìï¥ÏÑù:
- Ïù¥ Îç∞Ïù¥ÌÑ∞ÏÖãÏùÄ Ìù¨Í∑ÄÏßàÌôò/Î≥µÏû°Ìïú ÏºÄÏù¥Ïä§ Ï§ëÏã¨
- ÍµêÍ≥ºÏÑúÏ†ÅÏù∏ "Ï†ÑÌòïÏ†Å Ìå®ÌÑ¥"Î≥¥Îã§ Í∞úÎ≥Ñ ÌôòÏûê Í≥†Ïú† Î≥ëÎ¶¨Í∞Ä ÏßÄÎ∞∞Ï†Å
- Í∞úÏù∏Ìôî ÏùòÎ£å(Personalized Medicine)Ïùò ÌïÑÏöîÏÑ± ÏûÖÏ¶ù

‚ö†Ô∏è ÌïúÍ≥Ñ:
- ÏÉòÌîå Ïàò Î∂ÄÏ°± ({len(df_final)}Í±¥)ÏúºÎ°ú ÏùºÎ∞òÌôî Ïñ¥Î†§ÏõÄ
- Î∞îÏù¥ÌÉà ÏÇ¨Ïù∏(ÌòàÏïï, Îß•Î∞ï Îì±) Î∂ÄÏû¨Î°ú Ï§ëÏ¶ùÎèÑ ÌèâÍ∞Ä Ï†úÌïú
    """)
    
    return df_final


df_final = meaningful_clustering_analysis(df_final)


# =============================================================================
# PART 3: ÏùòÎ£åÏ†ÅÏúºÎ°ú ÏùòÎØ∏ÏûàÎäî Ïû¨ÏûÖÏõê Ï†ïÏùò
# =============================================================================

def calculate_clinically_meaningful_readmission(df_acpt, df_adcpt):
    """
    ÏùòÎ£å ÌëúÏ§ÄÏóê ÎßûÎäî Ïû¨ÏûÖÏõê Ï†ïÏùò:
    1. 30Ïùº ÎÇ¥ Ïû¨ÏûÖÏõê
    2. Í≥ÑÌöçÎêòÏßÄ ÏïäÏùÄ Ïû¨ÏûÖÏõê (ÎèôÏùº ÏßàÎ≥ë ÏãúÏä§ÌÖú)
    """
    print("\n" + "="*70)
    print("üè• ÏûÑÏÉÅÏ†ÅÏúºÎ°ú ÏùòÎØ∏ÏûàÎäî Ïû¨ÏûÖÏõê Î∂ÑÏÑù")
    print("="*70)
    
    # ÏßÑÎã® Ï†ïÎ≥¥ Î≥ëÌï©
    df = df_acpt.merge(df_adcpt[['PatientID', 'AdmissionID', 'PrimaryDiagnosisDescription']], 
                       on=['PatientID', 'AdmissionID'], how='left')
    
    # Body System Î∂ÑÎ•ò
    df['BodySystem'] = df['PrimaryDiagnosisDescription'].apply(classify_by_body_system_improved)
    
    # ÏãúÍ∞ÑÏàú Ï†ïÎ†¨
    df = df.sort_values(['PatientID', 'AdmissionStartDate']).reset_index(drop=True)
    
    # Ïù¥Ï†Ñ ÏûÖÏõê Ï†ïÎ≥¥
    df['PrevAdmissionEnd'] = df.groupby('PatientID')['AdmissionEndDate'].shift(1)
    df['PrevBodySystem'] = df.groupby('PatientID')['BodySystem'].shift(1)
    
    # Ïû¨ÏûÖÏõê Í∞ÑÍ≤© Í≥ÑÏÇ∞
    df['DaysSincePrevDischarge'] = (
        pd.to_datetime(df['AdmissionStartDate']) - pd.to_datetime(df['PrevAdmissionEnd'])
    ).dt.total_seconds() / 86400
    
    # 30Ïùº ÎÇ¥ Ïû¨ÏûÖÏõê (Ï†ÑÏ≤¥)
    df['Readmit30_Any'] = (
        (df['DaysSincePrevDischarge'] > 0) & 
        (df['DaysSincePrevDischarge'] <= 30)
    ).astype(int)
    
    # 30Ïùº ÎÇ¥ + Í∞ôÏùÄ Body System (Í¥ÄÎ†® Ïû¨ÏûÖÏõê)
    df['Readmit30_Related'] = (
        (df['DaysSincePrevDischarge'] > 0) & 
        (df['DaysSincePrevDischarge'] <= 30) &
        (df['BodySystem'] == df['PrevBodySystem'])
    ).astype(int)
    
    print(f"Ï†ÑÏ≤¥ ÏûÖÏõê: {len(df)}Í±¥")
    print(f"30Ïùº ÎÇ¥ Ïû¨ÏûÖÏõê (Ï†ÑÏ≤¥): {df['Readmit30_Any'].sum()}Í±¥ ({df['Readmit30_Any'].mean()*100:.1f}%)")
    print(f"30Ïùº ÎÇ¥ Í¥ÄÎ†® Ïû¨ÏûÖÏõê: {df['Readmit30_Related'].sum()}Í±¥ ({df['Readmit30_Related'].mean()*100:.1f}%)")
    
    # Ïû¨ÏûÖÏõê Î∂ÑÏÑù
    readmit_analysis = df[df['Readmit30_Any'] == 1][
        ['PatientID', 'AdmissionStartDate', 'BodySystem', 'PrevBodySystem', 'DaysSincePrevDischarge']
    ]
    
    if len(readmit_analysis) > 0:
        print(f"\nÏû¨ÏûÖÏõê ÏºÄÏù¥Ïä§ ÏÉòÌîå:")
        print(readmit_analysis.head(10))
    
    # ÏãúÍ∞ÅÌôî
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))
    
    # Ïû¨ÏûÖÏõê Í∞ÑÍ≤© Î∂ÑÌè¨
    readmit_cases = df[df['Readmit30_Any'] == 1]
    if len(readmit_cases) > 0:
        axes[0].hist(readmit_cases['DaysSincePrevDischarge'], bins=30, edgecolor='black', color='coral')
        axes[0].set_title('Ïû¨ÏûÖÏõê Í∞ÑÍ≤© Î∂ÑÌè¨ (30Ïùº Ïù¥ÎÇ¥)')
        axes[0].set_xlabel('Ìá¥Ïõê ÌõÑ ÏùºÏàò')
        axes[0].set_ylabel('ÎπàÎèÑ')
        axes[0].axvline(x=7, color='red', linestyle='--', label='1Ï£º (Í∏âÏÑ± Ìï©Î≥ëÏ¶ù)')
        axes[0].legend()
    else:
        axes[0].text(0.5, 0.5, 'Ïû¨ÏûÖÏõê ÏºÄÏù¥Ïä§ ÏóÜÏùå', ha='center', va='center', fontsize=14)
        axes[0].set_title('Ïû¨ÏûÖÏõê Í∞ÑÍ≤© Î∂ÑÌè¨')
    
    # Ïû¨ÏûÖÏõê Ïú†Ìòï ÎπÑÍµê
    readmit_data = {
        'Ï†ÑÏ≤¥ Ïû¨ÏûÖÏõê': df['Readmit30_Any'].sum(),
        'Í¥ÄÎ†® Ïû¨ÏûÖÏõê': df['Readmit30_Related'].sum(),
        'ÎπÑÍ¥ÄÎ†® Ïû¨ÏûÖÏõê': df['Readmit30_Any'].sum() - df['Readmit30_Related'].sum()
    }
    
    axes[1].bar(readmit_data.keys(), readmit_data.values(), 
                color=['steelblue', 'seagreen', 'orange'], alpha=0.7)
    axes[1].set_title('Ïû¨ÏûÖÏõê Ïú†Ìòï Î∂ÑÏÑù')
    axes[1].set_ylabel('Ïû¨ÏûÖÏõê Í±¥Ïàò')
    axes[1].tick_params(axis='x', rotation=15)
    
    for i, (k, v) in enumerate(readmit_data.items()):
        axes[1].text(i, v + 0.5, str(v), ha='center', fontsize=12)
    
    plt.tight_layout()
    plt.show()
    
    # Ïù∏ÏÇ¨Ïù¥Ìä∏
    total_readmit = df['Readmit30_Any'].sum()
    related_readmit = df['Readmit30_Related'].sum()
    
    print("\n" + "="*70)
    print("üí° Ïû¨ÏûÖÏõê Î∂ÑÏÑù Ïù∏ÏÇ¨Ïù¥Ìä∏")
    print("="*70)
    
    if total_readmit > 0:
        related_pct = related_readmit / total_readmit * 100
        unrelated_pct = 100 - related_pct
        
        print(f"""
‚úÖ Ï£ºÏöî Î∞úÍ≤¨:
- 30Ïùº ÎÇ¥ Ïû¨ÏûÖÏõêÎ•†: {df['Readmit30_Any'].mean()*100:.1f}%
- Ïù¥ Ï§ë Í∞ôÏùÄ ÏßàÎ≥ë ÏãúÏä§ÌÖú: {related_pct:.1f}%
- Îã§Î•∏ ÏßàÎ≥ë ÏãúÏä§ÌÖú: {unrelated_pct:.1f}%

üìä ÏûÑÏÉÅÏ†Å Ìï¥ÏÑù:
- Ïû¨ÏûÖÏõêÏùò {unrelated_pct:.0f}%Í∞Ä Ï†ÑÌòÄ Îã§Î•∏ ÏßàÎ≥ë ÏãúÏä§ÌÖúÏúºÎ°ú Ïû¨ÏûÖÏõê
- Ïù¥Îäî "ÎßåÏÑ± Î≥µÌï© ÏßàÌôòÏûê" ÌäπÏÑ± ÏãúÏÇ¨
- ÎèôÎ∞ò Ïù¥Ìôò(Comorbidity) Í¥ÄÎ¶¨Ïùò Ï§ëÏöîÏÑ±

‚ö†Ô∏è Î™®Îç∏ÎßÅ Ï†úÏïΩ:
- ÎπÑÍ¥ÄÎ†® Ïû¨ÏûÖÏõêÏùÄ Ïù¥Ï†Ñ ÏûÖÏõê Îç∞Ïù¥ÌÑ∞Î°ú ÏòàÏ∏° Î∂àÍ∞ÄÎä•
- Í¥ÄÎ†® Ïû¨ÏûÖÏõê ÏºÄÏù¥Ïä§Îßå {related_readmit}Í±¥ ‚Üí Î™®Îç∏ ÌïôÏäµ Î∂àÍ∞Ä
        """)
    else:
        print("""
‚ö†Ô∏è 30Ïùº ÎÇ¥ Ïû¨ÏûÖÏõê ÏºÄÏù¥Ïä§Í∞Ä ÏóÜÏäµÎãàÎã§.
- ÏÉòÌîå Í∏∞Í∞ÑÏù¥ ÏßßÍ±∞ÎÇò ÌôòÏûê Ï∂îÏ†ÅÏù¥ Î∂àÏôÑÏ†ÑÌï† Í∞ÄÎä•ÏÑ±
- Ïû¨ÏûÖÏõê ÏòàÏ∏° Î™®Îç∏ Íµ¨Ï∂ï Î∂àÍ∞Ä
        """)
    
    return df


readmission_df = calculate_clinically_meaningful_readmission(df_ACPT, df_ADCPT)


# =============================================================================
# PART 4: LOS ÏòàÏ∏° ÌÉÄÎãπÏÑ± Í≤ÄÏ¶ù
# =============================================================================

def validate_los_prediction_feasibility(df_final):
    """
    ÌòÑÏû¨ FeatureÎ°ú LOS ÏòàÏ∏°Ïù¥ Í∞ÄÎä•ÌïúÏßÄ Í≤ÄÏ¶ù
    """
    print("\n" + "="*70)
    print("üìä LOS ÏòàÏ∏° ÌÉÄÎãπÏÑ± Î∂ÑÏÑù")
    print("="*70)
    
    if 'LengthOfStay_Days' not in df_final.columns:
        print("‚ö†Ô∏è LengthOfStay_Days Ïª¨Îüº ÏóÜÏùå")
        return None
    
    # ÏÇ¨Ïö© Í∞ÄÎä•Ìïú Feature
    lab_cols = [c for c in df_final.columns if c.startswith('LabValue_')]
    missing_ratio = df_final[lab_cols].isnull().mean()
    valid_lab_cols = missing_ratio[missing_ratio < 0.5].index.tolist()
    
    print(f"ÏÇ¨Ïö© Í∞ÄÎä•Ìïú Í≤ÄÏÇ¨ Ìï≠Î™©: {len(valid_lab_cols)}Í∞ú")
    
    # Feature ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ Î∂ÑÏÑù
    feature_cols = valid_lab_cols + ['AgeAtAdmission']
    feature_cols = [c for c in feature_cols if c in df_final.columns]
    
    X = df_final[feature_cols].copy()
    y = df_final['LengthOfStay_Days'].copy()
    
    # Í≤∞Ï∏°Ïπò Ï†úÍ±∞
    valid_idx = X.notna().all(axis=1) & y.notna()
    X = X[valid_idx].fillna(X.median())
    y = y[valid_idx]
    
    # Í∞Å FeatureÏôÄ LOS Í∞Ñ ÏÉÅÍ¥ÄÍ≥ÑÏàò Í≥ÑÏÇ∞
    correlations = []
    for col in X.columns:
        corr, pval = stats.pearsonr(X[col], y)
        correlations.append({
            'Feature': col,
            'Correlation': corr,
            'P-value': pval
        })
    
    corr_df = pd.DataFrame(correlations).sort_values('Correlation', key=abs, ascending=False)
    
    print("\nÏÉÅÏúÑ 10Í∞ú FeatureÏôÄ LOS ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ:")
    print(corr_df.head(10))
    
    # Ïú†ÏùòÎØ∏Ìïú Feature Í∞úÏàò
    significant = corr_df[corr_df['P-value'] < 0.05]
    print(f"\nÌÜµÍ≥ÑÏ†ÅÏúºÎ°ú Ïú†ÏùòÌïú Feature: {len(significant)}Í∞ú (p < 0.05)")
    
    # ÏãúÍ∞ÅÌôî
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))
    
    # ÏÉÅÍ¥ÄÍ≥ÑÏàò Î∂ÑÌè¨
    axes[0].barh(corr_df.head(15)['Feature'], corr_df.head(15)['Correlation'], 
                 color=['red' if p < 0.05 else 'gray' for p in corr_df.head(15)['P-value']])
    axes[0].set_title('Feature-LOS ÏÉÅÍ¥ÄÍ≥ÑÏàò (Îπ®Í∞ï=Ïú†Ïùò, ÌöåÏÉâ=ÎπÑÏú†Ïùò)')
    axes[0].set_xlabel('Pearson Correlation')
    axes[0].axvline(x=0, color='black', linestyle='--', linewidth=1)
    
    # Ïã§Ï†ú ÏòàÏ∏° ÏãúÎèÑ
    if len(significant) >= 3:
        top_features = significant.head(10)['Feature'].tolist()
        X_model = X[top_features]
        
        X_train, X_test, y_train, y_test = train_test_split(
            X_model, y, test_size=0.3, random_state=42
        )
        
        rf = RandomForestRegressor(n_estimators=50, max_depth=5, random_state=42)
        rf.fit(X_train, y_train)
        y_pred = rf.predict(X_test)
        
        mae = mean_absolute_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)
        
        axes[1].scatter(y_test, y_pred, alpha=0.6, s=50)
        axes[1].plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 
                     'r--', lw=2, label='ÏôÑÎ≤ΩÌïú ÏòàÏ∏°')
        axes[1].set_title(f'LOS ÏòàÏ∏° Í≤∞Í≥º (R¬≤={r2:.3f}, MAE={mae:.2f}Ïùº)')
        axes[1].set_xlabel('Ïã§Ï†ú LOS (Ïùº)')
        axes[1].set_ylabel('ÏòàÏ∏° LOS (Ïùº)')
        axes[1].legend()
        axes[1].grid(True, alpha=0.3)
    else:
        axes[1].text(0.5, 0.5, 'Ïú†ÏùòÌïú Feature Î∂ÄÏ°±\nÏòàÏ∏° Î™®Îç∏ ÌïôÏäµ Î∂àÍ∞Ä', 
                    ha='center', va='center', fontsize=14)
        axes[1].set_title('LOS ÏòàÏ∏° ÏãúÎèÑ')
    
    plt.tight_layout()
    plt.show()
    
    # Ïù∏ÏÇ¨Ïù¥Ìä∏
    print("\n" + "="*70)
    print("üí° LOS ÏòàÏ∏° ÌÉÄÎãπÏÑ± Ïù∏ÏÇ¨Ïù¥Ìä∏")
    print("="*70)
    
    if len(significant) >= 3:
        print(f"""
‚úÖ ÏòàÏ∏° Í∞ÄÎä•ÏÑ±: Ï†úÌïúÏ†Å
- Ïú†ÏùòÌïú Feature: {len(significant)}Í∞ú
- ÏòàÏ∏° ÏÑ±Îä•: R¬≤={r2:.3f}, MAE={mae:.2f}Ïùº

üìä ÌïúÍ≥ÑÏ†ê:
- R¬≤ < 0.3: ÌòÑÏû¨ FeatureÎ°úÎäî LOS Î≥ÄÎèôÏùò ÎåÄÎ∂ÄÎ∂ÑÏùÑ ÏÑ§Î™Ö Î™ªÌï®
- Î∂ÄÏ°±Ìïú Feature:
  * Î∞îÏù¥ÌÉà ÏÇ¨Ïù∏ (ÌòàÏïï, Îß•Î∞ï, Ï≤¥Ïò®)
  * ÏßàÎ≥ë Ï§ëÏ¶ùÎèÑ Ï†êÏàò (APACHE, SOFA)
  * ÏπòÎ£å ÎÇ¥Ïó≠ (ÏàòÏà†, ÏïΩÎ¨º)
  * ÎèôÎ∞ò ÏßàÌôò Ï†ïÎ≥¥

üí° Í∞úÏÑ† Î∞©Ìñ•:
- Body SystemÎ≥Ñ Î∂ÑÎ¶¨ Î™®Îç∏ Í≥†Î†§
- ÏãúÍ≥ÑÏó¥ Feature (Í≤ÄÏÇ¨ ÏàòÏπò Î≥ÄÌôîÏú®) Ï∂îÍ∞Ä
- Ïô∏Î∂Ä Îç∞Ïù¥ÌÑ∞ Î≥¥Í∞ï ÌïÑÏöî
        """)
    else:
        print(f"""
‚ùå ÏòàÏ∏° Î∂àÍ∞Ä
- Ïú†ÏùòÌïú Feature: {len(significant)}Í∞ú (ÏµúÏÜå 3Í∞ú ÌïÑÏöî)
- ÌòÑÏû¨ Í≤ÄÏÇ¨ Ìï≠Î™©ÎßåÏúºÎ°úÎäî LOS ÏòàÏ∏° Î∂àÍ∞ÄÎä•

‚ö†Ô∏è Í∑ºÎ≥∏ ÏõêÏù∏:
- Í≤ÄÏÇ¨ Ìï≠Î™© Ïàò Î∂ÄÏ°±
- ÏßàÎ≥ëÏùò Îã§ÏñëÏÑ± (Ìù¨Í∑ÄÏßàÌôò Ï§ëÏã¨)
- ÏÉòÌîå Ïàò Î∂ÄÏ°± (Ìå®ÌÑ¥ ÌïôÏäµ Ïñ¥Î†§ÏõÄ)
        """)
    
    return corr_df


los_feasibility = validate_los_prediction_feasibility(df_final)


```
